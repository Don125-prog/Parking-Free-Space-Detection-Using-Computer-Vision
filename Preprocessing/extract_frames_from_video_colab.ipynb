```python
# Функция для извлечения кадров из видео на Google Colab
def extract_frames_from_video_colab(video_path, output_folder, name_images='frame_', interval_min=5, interval_max=10, sizeX = 640, sizeY = 640):

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        return

    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    video_length = total_frames / fps if fps > 0 else 0

    print(f"Загружено видео: {total_frames} кадров, длина видео: {video_length:.2f} сек.")

    current_time = 0
    saved_count = 0

    #Основной цикл
    while current_time < video_length:
        cap.set(cv2.CAP_PROP_POS_MSEC, current_time * 1000)  # Устанавливаем позицию видеозаписи в миллисекундах
        success, frame = cap.read()
        if not success:
            break

        resized = cv2.resize(frame, (sizeX, sizeY))
        filename = f"{name_images}{saved_count:04d}.jpg"
        filepath = os.path.join(output_dir, filename)
        cv2.imwrite(filepath, resized, [cv2.IMWRITE_JPEG_QUALITY, 95])

        print(f"Сохранил кадр {saved_count} на {current_time:.2f} с -> {filepath}")
        saved_count += 1

        current_time += random.uniform(interval_min, interval_max)

    cap.release()
    print(f"Итого сохранено: {saved_count} кадров в {output_dir}")

```

```python
video_path = "/content/parking.mp4"  # Замените на путь к вашему видео в Colab
output_dir = "/content/parking"  # Папка на Google Drive или локально
extract_frames_from_video_colab(video_path, output_dir)
```
