```python
import cv2
import os
import random
from google.colab import drive
from google.colab.patches import cv2_imshow
```

```python
if not os.path.exists("/content/video"):
  os.makedirs("/content/video")

if not os.path.exists("/content/parking"):
  os.makedirs("/content/parking")
```

```python
video_dir ="/content/video"
output_dir = "/content/parking"
```

```python
# Функция для извлечения кадров из видео на Google Colab
def extract_frames_from_video_colab(video_path, output_dir, saved_count = 0, interval_min=1, interval_max=3, sizeX = 640, sizeY = 640):

    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        return 0

    fps = cap.get(cv2.CAP_PROP_FPS)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    video_length = total_frames / fps if fps > 0 else 0


    current_time = 0

    #Основной цикл
    while current_time < video_length:
        cap.set(cv2.CAP_PROP_POS_MSEC, current_time * 1000)  # Устанавливаем позицию видеозаписи в миллисекундах
        success, frame = cap.read()
        if not success:
            break

        resized = cv2.resize(frame, (sizeX, sizeY))
        filename = f"{saved_count:04d}.jpg"
        filepath = os.path.join(output_dir, filename)
        cv2.imwrite(filepath, resized, [cv2.IMWRITE_JPEG_QUALITY, 95])

        saved_count += 1

        current_time += random.uniform(interval_min, interval_max)

    cap.release()
    return saved_count
```

```python
saved_count = 0
```

```python
for fname in os.listdir(video_dir):
    if not fname.lower().endswith(".mp4"):
        continue
    video_path = os.path.join(video_dir, fname)
    saved_count = extract_frames_from_video_colab(video_path, output_dir, saved_count)
```
